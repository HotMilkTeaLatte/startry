```python
# 상권
data_sangju_0 = pd.read_csv('상권_상주인구.csv')
data_saenghwal_0 = pd.read_csv('상권_생활인구.csv')
data_sodeugsobi_0 = pd.read_csv('상권_소득소비.csv')
data_jeompo_0 = pd.read_csv('상권_점포.csv')
data_jigjang_0 = pd.read_csv('상권_직장인구.csv')
data_jibgaeg_0 = pd.read_csv('상권_집객시설.csv')
data_maechul_0 = pd.read_csv('상권_추정매출.csv')
data_jipyo_0 = pd.read_csv('상권_상권변화지표.csv')

# 행정동 단위 외국인 생활인구
data_dangi = pd.read_csv('행정동_단위_서울_생활인구(단기체류_외국인).csv')
data_janggi = pd.read_csv('행정동_단위_서울_생활인구(장기체류_외국인).csv')
```

```python
print('--------상권 상주인구----------')
print(data_sangju_0.isnull().sum())      # 결측치 없음
print('--------상권 생활인구----------')
print(data_saenghwal_0.isnull().sum())      # 결측치 없음
print('--------상권 소득소비----------')
print(data_sodeugsobi_0.isnull().sum())      # 결측치 있음 
print('--------상권 점포--------------')
print(data_jeompo_0.isnull().sum())      # 결측치 없음
print('--------상권 직장인구----------')
print(data_jigjang_0.isnull().sum())      # 결측치 없음
print('--------상권 집객시설----------')
print(data_jibgaeg_0.isnull().sum())      # 결측치 없음
print('--------상권 추정매출----------')
print(data_maechul_0.isnull().sum())      # 결측치 없음
print('--------상권 변화지표----------')
print(data_jipyo_0.isnull().sum())      # 결측치 없음
```
